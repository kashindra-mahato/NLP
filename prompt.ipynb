```python
from haystack import Pipeline
from haystack.nodes import TextConverter, PreProcessor

indexing_pipeline = Pipeline()
text_converter = TextConverter()
preprocessor = PreProcessor(
    clean_whitespace=True,
    clean_header_footer=True,
    clean_empty_lines=False,
    split_by="sentence",
    split_length=1,
    # split_overlap=20,
    split_respect_sentence_boundary=False,
)
```

```python
import openai
openai.api_key = ""
def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0.2, # this is the degree of randomness of the model's output
    )
#     print(str(response.choices[0].message))
    return response.choices[0].message["content"]
```

```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.get("choices")[0].get("message").get("content"), response.get("usage")
```

```python
import csv
from datetime import datetime
def query(test_data, pipe):
  for i in range(len(test_data['Queries '])):
    query = test_data['Queries '][i]
    documents = pipe.run(query=query, params={"retriever":{"top_k": 3}})
    paragraph = ", ".join([documents["documents"][j].content for j in range(len(documents["documents"]))])
    # print(f"##################################\n{paragraph}################################")
    prompt = f"""
            I want you to act as a foreign studies assistant bot for ----.\
            I will provide you with a pair of Query: {query} and Response: {paragraph}\
            and you will act as the bot’s research assistant.\
            First extract exact answer of Query from Response.\
            If response does not provide answer to the query say 'answer not found'\
            do not add anything before or after 'answer not found'\
            then rephrase the answer in clear and concise way\
            Finally only provide answer in less than 30 words as\
            Answer:
            """
    if len(paragraph) > 0:
      response, count = get_completion(prompt)
      time = datetime.now()
      completion_tokens = count.get("completion_tokens")
      prompt_tokens = count.get("prompt_tokens")
      total_tokens = count.get("total_tokens")
    else:
      response = "Sorry, not found."
    print(f"\n\nQuery: {query}\n{response}\n\n")

    # Create the dictionary (=row)
    row = {'Query':query,'Response':response,'completion_tokens':completion_tokens, 'prompt_tokens':prompt_tokens, 'total_tokens':total_tokens, 'time':time}
    # Open the CSV file in "append" mode
    with open('.csv', 'a', newline='') as f:
        # Create a dictionary writer with the dict keys as column fieldnames
        writer = csv.DictWriter(f, fieldnames=row.keys())
        # Append single row to CSV
        writer.writerow(row)
```
